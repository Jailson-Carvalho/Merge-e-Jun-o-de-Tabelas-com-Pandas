### Informações em Banco de Dados diferente

#### Pode haver erros se a query for feita no SQL, então faremos aqui.

```python
import pandas as pd 
import pyodbc


server = 'DESKTOP-TDTQ2ED' # substitua pelo nome do servidor SQL Server
database = 'Python' # substitua pelo nome do BD
conexaoBD = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};'# driver do SQL Server 
                    f'SERVER={server};'
                    f'DATABASE={database};'
                    'Trusted_Connection=yes;')

cursor = conexaoBD.cursor() # O cursor é um objeto de controle que faz a ponte entre o Python e o banco de dados.
```

#### Query para fazer a importação

```python
items ="""

     SELECT * FROM ITEMS WHERE total_price >= 1000

"""
```

```python
query1 = pd.read_sql(items, conexaoBD)
query1

```

### Fazer a conexão com o segundo Banco de Dados

```python
import pandas as pd 
import pyodbc


server2 = 'DESKTOP-TDTQ2ED' # substitua pelo nome do servidor SQL Server
database2 = 'ContosoRetailDW' # substitua pelo nome do BD
conexaoBD2 = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};'# driver do SQL Server 
                    f'SERVER={server2};'
                    f'DATABASE={database2};'
                    'Trusted_Connection=yes;')

cursor2 = conexaoBD2.cursor() # O cursor é um objeto de controle que faz a ponte entre o Python e o banco de dados.
```

### Criar a conexão com o segundo Banco de Dados

```python
query = """

    SELECT 

	ProductKey,
	ProductName,
	UnitPrice

FROM DimProduct

"""
```

```python
product = pd.read_sql(query, conexaoBD2)

product
```

###  Vamos fazer um merge/juntar

```python
join = pd.merge(query1, product, left_on='product_id', right_on='ProductKey', how='left')
join
```
